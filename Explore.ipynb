```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
df = pd.DataFrame(data.data, 
                  columns=['sl', 'sw', 'pl', 'pw'])
df['Species'] = data.target
```

```python
df = df[df['Species'] != 1]
```

```python
sns.scatterplot(x='pl', y='pw', data=df, hue='Species')
plt.show()
```
